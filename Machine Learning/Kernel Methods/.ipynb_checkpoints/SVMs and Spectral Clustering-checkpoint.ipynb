# 1. SVM with randomly selected labels

The first result is to analize the changes in performance of the SVM as the number of available labelled examples increases. This problem consists in to separate 0s from 1s.



The code in the following cell loads these data and plots some of the examples.

```python
Y = np.loadtxt("y_01.csv", 
                       delimiter=",") 
X = np.loadtxt("x_01.csv", 
                       delimiter=",")

nrows = 4
ncols = 4
ff,aa = plt.subplots(nrows, ncols,figsize=(4,4))
ii = 0
for rr in range(nrows):
    for cc in range(ncols):
        aa[rr][cc].set_axis_off()
        aa[rr][cc].imshow(X[ii].reshape(28,28), 
                          cmap=plt.cm.gray_r, 
                          interpolation='nearest')
        aa[rr][cc].set_title(repr(Y[ii]))
        ii += 1
ff.tight_layout()  
```

It seems a reasonably easy classification in terms of the separability of the classes. Therefore, it should be easy to achieve good classification results with a training dataset of relative small size. 

We will now analyze how the results of the SVM improve as we get more and more labelled data.

```python
v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400]) # number of labelled samples
acc_svm_random_label1 = np.empty(len(v_num_labels))
C = 100
gamma = 0.001

for nn in range(len(v_num_labels)):
    num_data = v_num_labels[nn]
    idx = np.random.randint(X.shape[0],size=(1,num_data))[0]
    predictors = X[idx,:]
    labels = Y[idx]
    
    if len(np.unique(labels)) == 1:
        acc_svm_random_label1[nn] = 0.5
        
    else:
        model = SVC(gamma=gamma,C=C,kernel='rbf',random_state=nn).fit(predictors, labels)
        acc_svm_random_label1[nn] = model.score(X,Y)
    
        
plt.figure()

plt.plot(v_num_labels, acc_svm_random_label1, label='SVM, labels got at random')

plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

Now getting a more robust statistical significance of the results by running the above loop a number of `NSIM` times.

```python
v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400])
NSIM=100
acc_svm_random_label_0vs1 = np.empty((NSIM, len(v_num_labels)))

for j in range(NSIM):
    for nn in range(len(v_num_labels)):
        num_data = v_num_labels[nn]
        idx = np.random.randint(X.shape[0],size=(1,num_data))[0]
        predictors = X[idx,:]
        labels = Y[idx]

        if len(np.unique(labels)) == 1:
            acc_svm_random_label_0vs1[j,nn] = 0.5

        else:
            model = SVC(gamma=gamma,C=C,kernel='rbf',random_state=j*nn).fit(predictors, labels)
            acc_svm_random_label_0vs1[j,nn] = model.score(X,Y)

plt.figure()

plt.plot(v_num_labels, np.mean(acc_svm_random_label_0vs1,0), label='SVM, labels got at random', color='blue')
plt.fill_between(v_num_labels, np.percentile(acc_svm_random_label_0vs1,25,axis=0),
                 np.percentile(acc_svm_random_label_0vs1,75, axis=0),
                 color='blue',alpha=0.3)
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

We can clearly see how as the number of available **labels** to train the SVM **increases**, the **uncertainty decreases**.

With very few instances (2-6), the uncertainty is very large, ranging from 0.5 to 0.9 in its highest point of uncertainty.

However, we can see that with only **10 instances** available, the uncertainty is already quite low and we achieve a very good performance (**above 0.95**).

This is very reasonable because if we fit the model with very few samples (only 2), it could either happen that both observations are very good to distinguish between 0's and 1's (both very representative of their own group); or that they are very misleading (they belong to the same class, or the quality of the drawing is poor); of course, anything in between is also possible.  
Thus, in the first case we would get a high accuracy (close to 0.9), a bad one in the second case (close to 0.5), and the rest of cases range in between those values.

# 2.- Spectral Clustering with one label per cluster

```python
from sklearn.cluster import SpectralClustering
from sklearn.neighbors import KNeighborsClassifier
import random

class oracle(object):

    def __init__(self, x, y):
        self.dim = x.shape[1]
        self.knn = KNeighborsClassifier(n_neighbors=1).fit(x,y)
        
    def get_label(self, x):
        return self.knn.predict(x.reshape(-1,self.dim))
    
def my_spectral_clustering(x, my_oracle, n_clusters=2, gamma=1):
    Nt = x.shape[0]
    micluster = SpectralClustering(n_clusters=n_clusters,gamma=gamma,random_state=100463947).fit(x) # set seed for reproducibility
    predictions = np.empty(Nt)
    
    for cc in range(n_clusters):
        my_x = random.choice(x[micluster.labels_ == cc])
        y_cluster = my_oracle.get_label(my_x)
        predictions[micluster.labels_ == cc] = y_cluster
        
    return predictions
```

Now use the function `my_spectral_clustering` to find out the performance of this proposed classification method based in clustering with a reduced number of labels using the number of clusters indicated by `v_num_labels` and a value of `gamma=0.001` for the RBF kernel of the spectral clustering. 

```python
v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400])
acc_spectral_clustering = np.empty(len(v_num_labels))
gamma = 0.001
my_oracle = oracle(X,Y)
for ii,n_clusters in enumerate(v_num_labels):
    predictions = my_spectral_clustering(X, my_oracle, n_clusters=n_clusters, gamma=gamma)
    acc_spectral_clustering[ii] = np.mean(predictions == Y)
```

Plotting the classification accuracy of the spectral clustering as a function of the number of available labels

```python
plt.figure()

plt.plot(v_num_labels, acc_spectral_clustering, label='SVM, labels got at random')

plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

Exploring the dependence of the classification accuracy of the spectral clustering with the values of `gamma`. 

```python
v_gamma = [1e-4, 1e-3, 1e-2, 1e-1, 5e-1] 

acc_spectral_clustering = np.empty((len(v_gamma),len(v_num_labels)))

for j in range(len(v_gamma)):
    for k in range(len(v_num_labels)):
        predictions = my_spectral_clustering(X, my_oracle, n_clusters=v_num_labels[k], gamma=v_gamma[j])
        acc_spectral_clustering[j,k] = np.mean(predictions == Y)
```

```python
plt.figure()
for ig, gamma in enumerate(v_gamma[:-1]):
    plt.plot(v_num_labels, acc_spectral_clustering[ig,:], label='Spectral Clustering, gamma {0:.4f}'.format(gamma))
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

```python
# For 1e-1, we get much worse results; so remove it to get a better visual plot
v_gamma = [1e-4, 1e-3, 1e-2, 1e-1] 

acc_spectral_clustering_0vs1_1 = np.empty((len(v_gamma),len(v_num_labels)))

for j in range(len(v_gamma)):
    for k in range(len(v_num_labels)):
        predictions = my_spectral_clustering(X, my_oracle, n_clusters=v_num_labels[k], gamma=v_gamma[j])
        acc_spectral_clustering_0vs1_1[j,k] = np.mean(predictions == Y)
        
plt.figure()
for ig, gamma in enumerate(v_gamma[:-1]):
    plt.plot(v_num_labels, acc_spectral_clustering_0vs1_1[ig,:], label='Spectral Clustering, gamma {0:.4f}'.format(gamma))
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

I select **gamma to 0.01** because it performs the best when considering a relatively small number of labels (less than 100), which may be precisely the point of doing an unsupervised algorithm with a few labels to perform a supervised task.  

As we can see, once you have a more decent amount of labels (more than 100 it seems), the unsupervised algorithm performs relatively worse.  
This could happen because when there are many clusters, there are also few observations per cluster, so the probability that an observation wrongly assigned to that cluster will be selected as the labeled one increases.

For instance, if we have 200 clusters, there may be around 5 observations per cluster.  
Then, it is likely that for some clusters, there are observations corresponding to 0's and 1's for a given cluster.  
Therefore, as we would only have around 5 observations, it is likely that one of the wrong ones is selected, thus decreasing the accuracy, as the other observations from that cluster will also be wrongly predicted.

```python
v_num_labels = np.array([2,4,8,10,20,30,40,60,80,100,200,300,400])
gamma = 0.01
NSIM = 100
acc_spectral_clustering_0vs1_2 = np.empty((NSIM, len(v_num_labels)))

for j in range(NSIM):
  for ii,n_clusters in enumerate(v_num_labels):
      predictions = my_spectral_clustering(X, my_oracle, n_clusters=n_clusters, gamma=gamma)
      acc_spectral_clustering_0vs1_2[j,ii] = np.mean(predictions == Y)
```

```python
plt.figure()

plt.plot(v_num_labels, np.mean(acc_spectral_clustering_0vs1_2,0), label='Spectral Clustering, gamma 0.01',color='orange')
plt.fill_between(v_num_labels, np.percentile(acc_spectral_clustering_0vs1_2,25,axis=0),
                 np.percentile(acc_spectral_clustering_0vs1_2,75, axis=0),
                 color='orange',alpha=0.3)
plt.plot(v_num_labels, np.mean(acc_svm_random_label_0vs1,0), label='SVM, labels got at random', color='blue')
plt.fill_between(v_num_labels, np.percentile(acc_svm_random_label_0vs1,25,axis=0),
                 np.percentile(acc_svm_random_label_0vs1,75, axis=0),
                 color='blue',alpha=0.3)

plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

We can see two different behaviors between the two models depending on the number of labels.  

On the **SVM**, the number of labels indicates how many observations are used to train the model.  
This is why we see very large confidence intervals for low numbers, and very narrow confidence intervals (approximating to one) for larger numbers.

As for the **Spectral Clustering** model, the number of labels corresponds to the number of clusters the observations will be divided on.  
This actually results in the number of observations for which we need to know its label, as we use one label per cluster to determine the labels of the rest of observations of a certain cluster.  

We see an different behavior, with very similar results for low number of clusters, then a little more variability as the number of clusters grow.   
This low variablity across all the graph may be because as there are only 2 types of number to distinguish from, the clustering construction will be done similarly, thus resulting in repetetive accuracy scores.

For instance, 2 clusters will always distinguish between 0's and 1's in a broad way, 4 clusters may additionally distinguish narrow 0's, narrow 1's, and so on.  
At the end of the day, there doesn't seem to be that much variablity in the drawings of the number so that the cluster construction varies a lot.

Now, comparing the general performance of both models, we can see that the **SVM** reaches an accuracy of 1 from a number of labels equal to **30 onwards**.   
This makes sense, simply because as the number of labeled observations used to train a supervised model increases, the accuracy should also increase.

On the other hand, the **Spectral Clustering** starts with an accuracy very close to 1 until using around **60 clusters**, where the accuracy drops, then it increases again when using more than **200 clusters**.  
The intuition behind this behavior was previously explained.  
Additionally, one could think that past a certain point, we could have a number of clusters so large that almost every cluster has 1 or 2 observations, thus resulting in an increase in performance again.  
However, it could happen that for that many clusters, many of them are wrongly assigned, which would result in a drop in performance.  

We will check if this is correct on the next cell (takes a lot to run, so I will limit the values).

```python
v_num_labels_check = np.array([300,500,700,900])
gamma = 0.01
NSIM = 20
check = np.empty((NSIM, len(v_num_labels_check)))

for j in range(NSIM): 
    if j%5 == 0:
        print(j)
    for ii,n_clusters in enumerate(v_num_labels_check):
        predictions = my_spectral_clustering(X, my_oracle, n_clusters=n_clusters, gamma=gamma)
        check[j,ii] = np.mean(predictions == Y)
```

```python
plt.figure()

plt.plot(v_num_labels_check, np.mean(check,0), label='Spectral Clustering, gamma 0.01',color='orange')
plt.fill_between(v_num_labels_check, np.percentile(check,25,axis=0),
                 np.percentile(check,75, axis=0),
                 color='orange',alpha=0.3)

plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

As we can see, we eventually reach an accuracy of 100% because we have many clusters with only 1 observation, which will be selected and for which we will know its label.

In conclusion, according to these results, we can say that if we had **less than 20 labels**, the **Spectal Clustering** would be a better model.  
If the number of labels was **between 20 and 60**, **both models** would perform relatively equal, with an accuracy very close to 1.  
Finally, if we had **more than 60 labels** available, the **SVM** model would be a better option to take.

The problem now is still a binary classification problem, but we need to differentiate between odd and even numbers.  
The difference with the previous problem is that before, we only had to distinguish between 0's and 1's, where as now, we need to distinguish between **{0,2,4,6,8}** and **{1,3,5,7,9}**.

This task is **harder** to manage, as there are **more similarities** between the two groups (8 & 9, 6 & 9, 4 & 7, ...).  
Additionally, considering we would be using a small number labels, the problem becomes even harder, as we would have even **fewer examples of each number**.  
Finally, as the ways of drawing the digits increases, there could be more **misleading observations** (4's,7's,5's that look like 1's, etc).

Comparing first both SVM models, we see that as the second problem is a more complex one, it takes longer (more labeled observations) to achieve a good performance.  
We also see the uncertainty in the results behaving in a similar way (more uncertainty as the number of observations decreases).

Now, comparing both Spectral Clustering models, we can remark that for the second problem, the uncertainty is much larger for a small number of clusters.  
This makes sense, as we now have to aggregate 10 different digits into 2 groups.  
This is a much harder task to achieve. Thus, we could imagine that when aggregating all these observations in a small number of clusters (less than 10), the way in which it would construct these clusters would vary a lot. 

For instance, with 3 clusters, it could aggregate  
{0,6,8,9}, {1,7,5} and {2,3,4}  
but it would also be reasonable to think that it could do  
{1,2,7}, {0,9} and {3,4,5,6,8},  
depending on the way the numbers are drawn.

As for the selection of the best model, we have arrived to the following conclusions:

**0 vs 1 problem**  
- 2-20 labels: Spectral Clustering  
- 20-40 labels: Either model  
- 60-400 labels: SVM  


**Even vs odd problem**  
- 2-30 labels: Spectral Clustering  
- 30-400 labels: SVM  

# Extra 1

Using KMeans instead of Spectral Clustering and discussing for which values of `gamma` the Spectral Clustering is closer to KMeans.

```python
class oracle(object):
    
    def __init__(self, x, y):
        self.dim = x.shape[1]
        self.knn = KNeighborsClassifier(n_neighbors=1).fit(x,y)
        
    def get_label(self, x):
        return self.knn.predict(x.reshape(-1,self.dim))
    
def my_spectral_clustering(x, my_oracle, n_clusters=2, gamma=1):

    Nt = x.shape[0]
    micluster = SpectralClustering(n_clusters=n_clusters,gamma=gamma,random_state=100463947).fit(x) 
    predictions = np.empty(Nt)
    
    for cc in range(n_clusters):
        my_x = random.choice(x[micluster.labels_ == cc])
        y_cluster = my_oracle.get_label(my_x)
        predictions[micluster.labels_ == cc] = y_cluster
        
    return predictions

def my_kmeans(x, my_oracle, n_clusters=2):

    Nt = x.shape[0]
    micluster = KMeans(n_clusters=n_clusters,random_state=100463947).fit(x)
    predictions = np.empty(Nt) 
    
    for cc in range(n_clusters):
        my_x = random.choice(x[micluster.labels_ == cc])
        y_cluster = my_oracle.get_label(my_x)
        predictions[micluster.labels_ == cc] = y_cluster
        
    return predictions
```

```python
from sklearn.cluster import KMeans
v_gamma = [1e-4, 1e-3, 1e-2, 1e-1] 
my_oracle = oracle(X,Y)
acc_kmeans = np.empty(len(v_num_labels))

for k in range(len(v_num_labels)):
    predictions = my_kmeans(X, my_oracle, n_clusters=v_num_labels[k])
    acc_kmeans[k] = np.mean(predictions == Y)
        
plt.figure()

for ig, gamma in enumerate(v_gamma[:-1]):
    plt.plot(v_num_labels, acc_spectral_clustering_evenvsodd_1[ig,:], label='Spectral Clustering, gamma {0:.4f}'.format(gamma))
    
plt.plot(v_num_labels,acc_kmeans, label='KMeans') 

plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

If we analyze the performance of KMeans, we see that the accuracy stays between 0.7 and 0.8 when the number of labels is between 10 and 100.  
The model that is closer to this behavior is the one with **gamma** equal to **0.001**.  

Then, using more than 100 labels seems to make the predictions only better.  
The model that **least** follows this pattern is the one with **gamma** equal to **0.0001**, and perhaps the one that follows it the most is the model with **gamma** equal to **0.01** (but 0.001 is not that far off). 

Since the implementation of these methodologies would be to use the least amount of labels possible, I will consider with a heavier weight the first observation, and hence conclude that the model which is closer to KMeans is the one with **gamma** equal to **0.001**.

Checking what happens if you train the SVMs with the labelled samples that the spectral clustering uses to label each cluster instead of with the randomly selected labels.

```python
class oracle(object):
    
    def __init__(self, x, y):
        self.dim = x.shape[1]
        self.knn = KNeighborsClassifier(n_neighbors=1).fit(x,y)
        
    def get_label(self, x):
        return self.knn.predict(x.reshape(-1,self.dim))
    
def my_spectral_clustering(x, my_oracle, n_clusters=2, gamma=1):
    Nt = x.shape[0]
    micluster = SpectralClustering(n_clusters=n_clusters,gamma=gamma,random_state=100463947).fit(x) 
    predictions = np.empty(Nt)
    
    for cc in range(n_clusters):
        my_x = random.choice(x[micluster.labels_ == cc])
        y_cluster = my_oracle.get_label(my_x)
        predictions[micluster.labels_ == cc] = y_cluster
        
    return predictions

def mixed_model(x, n_clusters=2, gamma=1):
    micluster = SpectralClustering(n_clusters=n_clusters,gamma=gamma,random_state=100463947).fit(x)
    idx = []
    for cc in range(n_clusters):
        random_idx = random.sample(range(0,x[micluster.labels_ == cc].shape[0]),k=1)
        idx.extend(random_idx)
        
    predictors = X[idx,:]
    labels = Y[idx]
    
    if len(np.unique(labels)) == 1:
        return 0.5
    else:    
        model = SVC(gamma=gamma,C=C,kernel='rbf',random_state=nn).fit(predictors, labels)
        return model.score(X,Y)
```

```python
acc_mix = np.empty((len(v_gamma),len(v_num_labels)))

for j in range(len(v_gamma)):
    for k in range(len(v_num_labels)):
        acc_mix[j,k] = mixed_model(X, n_clusters=v_num_labels[k], gamma=v_gamma[j])
```

```python
plt.figure()
for ig, gamma in enumerate(v_gamma[:-1]):
    plt.plot(v_num_labels, acc_mix[ig,:], label='Mixed model, gamma {0:.4f}'.format(gamma))
    
plt.legend()
plt.xlabel('Num labels')
plt.ylabel('accuracy')
plt.xscale('log')
plt.grid()
```

First, comparing these results with the original SVM model, we see that the **performance is much worse**, and we **lose** the behavior that the accuracy increases as the number of labels increases, **converging to 1**.  

In fact, we reach a point in which the performance drops significantly (after around 200 labels) whereas before, it converged to an accuracy of 100%.

This **bias** when selecting the training observations has this effect because when using multiple observations, we have clusters with very few observations, and it seems like it tends to select a lot more observations from one class than from the other.  
Therefore, when the SVM is trained with these observations, it will most likely only **predict the "majority class"** among the labeled observations, hence resulting in this big drop in accuracy.
